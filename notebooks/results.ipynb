```python
import seml
import pandas as pd
pd.set_option('display.max_rows', None)
```

```python
results = seml.get_results('nodedp', to_data_frame=True).fillna(0)
```

```python
results.columns
```

```python
results = results.rename(columns={
    'config.model_args.arch': 'arch',
    'config.model_args.num_layers': 'layers', 
    'config.sampler_args.depth': 'depth', 
    'config.sampler_args.method': 'sampler', 
    'config.sampler_args.num_root': 'roots',
    'config.sampler_args.only_roots': 'only_roots',
    'config.training_args.alpha': 'alpha', 
    'config.training_args.delta': 'delta', 
    'config.training_args.method': 'trainer'
})
```

```python
col = ['arch', 'layers', 'depth', 'sampler', 'roots', 'only_roots', 'alpha', 'delta', 'trainer']
res_col = ['result.Accuracy',
       'result.iterations', 'result.gho', 'result.C',
       'result.gamma', 'result.eps', 'result.gho']
```

```python
results.groupby(col)[res_col].agg('mean')
```
